```python
# import dependancies
import matplotlib.pyplot as plt 
```

```python
# set the x-axis to a list of strings for each month
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# set the y-axis to a list of strings for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create plot and add a label for the legend
plt.plot(x_axis,y_axis,label = 'Boston')

#create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# set limit between 0 and 45
plt.ylim(0,45)

# create a title
plt.title('Pyber Fare by Month')

# add the legend
plt.legend()
```

```python
# create plot and add a label for the legend
plt.plot(x_axis,y_axis, marker="*",color = "green", linewidth=2, label = 'Boston')

#create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# set limit between 0 and 45
plt.ylim(0,45)

# create a title
plt.title('Pyber Fare by Month')

# add the legend
plt.legend()
plt.grid()
plt.show()
```

```python
# set x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# set y_axis to list of floats as the total fare in US Dollars per month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.barh(x_axis,y_axis,color='green',label='Boston')

# inverting y_axis
plt.gca().invert_yaxis()

# create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# create a title
plt.title("Pyber Fare by Month")

# add the legend
plt.legend()


plt.show()
```

```python
# set x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color = "Cyan",label="Chicago")
plt.legend()
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.show()
```

```python
plt.scatter(y_axis,x_axis, c="red")
plt.legend('Chicago')
plt.xlabel('rides')
plt.ylabel('months')
plt.xlim([0,50])
plt.show()
```

```python
plt.scatter(x_axis,y_axis,s=y_axis)
plt.show()
```

```python
y_axis_larger = []
for data in y_axis:
        y_axis_larger.append(data * 3)
plt.scatter(x_axis,y_axis, s=y_axis_larger)
plt.show()
```
